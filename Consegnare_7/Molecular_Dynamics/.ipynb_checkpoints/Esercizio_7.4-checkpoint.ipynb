```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
import numpy as np

N = 600 #t_max
x1 = np.arange(N)
z = np.loadtxt("output.temp.0", unpack='true', usecols=(1))
z_err = np.loadtxt("output.temp.0", unpack='true', usecols=(3))

plt.figure(figsize=(20,10))
plt.errorbar(x1, z, yerr=z_err)
plt.title('Temperatura')
plt.xlabel('Blocchi')
plt.grid(True)
plt.show();
```
