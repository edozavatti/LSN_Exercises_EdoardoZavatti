```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

M = 10000
N = 101
k = 1

x = np.arange(N)
y = np.loadtxt("media_random_walk.dat", unpack='true')
error = np.loadtxt("errore_random_walk.dat", unpack='true')

plt.axes([0., 0., 1.5, 1.5])
plt.errorbar(x, y, yerr=error)
plt.xlabel('Passi')
plt.ylabel('sqrt(<(r_N)^2>')
plt.title('Random walk discreto')
plt.grid(True) 

p_opt, p_cov = curve_fit(f, x, y, bounds=([0],[2]))
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit, color = "red") # plotting fitted function

print("Valor medio della distanza dall'origine al passo 100: ", y[99], " pm " , error[99])
print()
print("optimized parameters [k] =", p_opt)
print("parameters uncertainty =", np.sqrt(np.diagonal(p_cov)))
print()


plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

M = 10000
N = 101
k = 1

z = np.zeros(N)
x = np.arange(N)
y = np.loadtxt("media_random_walk.dat", unpack='true')
error = np.loadtxt("errore_random_walk.dat", unpack='true')

p_opt, p_cov = curve_fit(f, x, y, bounds=([0],[2]))

plt.axes([0., 0., 1.5, 1.5])
plt.errorbar(x, y - p_opt[0] * np.sqrt(x), yerr=error, label="simulated pi")
plt.plot(x, z, color = "red")
plt.xlabel('Passi')
plt.ylabel('sqrt(<(r_N)^2> - k * sqrt(passi)')
plt.title('Random walk discreto: verifica compatibilità')
plt.grid(True)


plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

M = 10000
N = 100
k = 1

x = np.arange(N)
y = np.loadtxt("media_random_walk_continuo.dat", unpack='true')
error = np.loadtxt("errore_random_walk_continuo.dat", unpack='true')

plt.axes([0., 0., 1.5, 1.5])
plt.errorbar(x, y, yerr=error)
plt.xlabel('Passi')
plt.ylabel('sqrt(<(r_N)^2>')
plt.title('Random walk continuo')
plt.grid(True)

p_opt, p_cov = curve_fit(f, x, y, bounds=([0.8],[1.2]))
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit, color = "red") # plotting fitted function

print("Valor medio della distanza dall'origine al passo 100: ", y[N-1], " pm " , error[N-1])
print()
print("optimized parameters [k] =", p_opt)
print("parameters uncertainty =", np.sqrt(np.diagonal(p_cov)))

plt.show();
```
