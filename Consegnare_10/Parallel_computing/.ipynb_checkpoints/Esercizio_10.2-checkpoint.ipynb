```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt

x1 = np.loadtxt("Route_quad_400", unpack='true', usecols=0)
y1 = np.loadtxt("Route_quad_400", unpack='true', usecols=1)

plt.plot(x1, y1)

plt.title('Initial path')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt

x1 = np.loadtxt("route_quad_0", unpack='true', usecols=0)
y1 = np.loadtxt("route_quad_0", unpack='true', usecols=1)
x2 = np.loadtxt("route_quad_1", unpack='true', usecols=0)
y2 = np.loadtxt("route_quad_1", unpack='true', usecols=1)
x3 = np.loadtxt("route_quad_2", unpack='true', usecols=0)
y3 = np.loadtxt("route_quad_2", unpack='true', usecols=1)

fig, axs = plt.subplots(1, 3)

#axs[0].figure(figsize=(7,5))
axs[0].plot(x1, y1)
#axs[1].figure(figsize=(7,5))
axs[1].plot(x2, y2)
axs[2].plot(x3, y3)

axs[1].set_title('Initial path')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```
