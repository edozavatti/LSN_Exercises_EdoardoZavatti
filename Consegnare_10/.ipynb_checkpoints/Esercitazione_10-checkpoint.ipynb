### Esercizio 10.1

Lo scopo del primo esercizio è di risolvere il TSP, nelle stesse condizioni dell'esercitazione 9 (città lungo un cerchio e dentro un quadrato) con il metodo del simulated annealing. Questa è una tecnica che, al pari degli algoritmi genetici, prende spunto dal funzionamento di sistemi fisici per risolvere problemi di ottimizzazione. In particolare l'osservazione chiave alla base del simulated annealing è che un sistema termodinamico liquido, se raffreddato lentamente, tende a formare un cristallo. Questo rappresenta uno stato di minima energia per il sistema, e viene così raggiunto in modo naturale qualora il raffreddamento sia lento.

Dunque, possiamo circoscrivere ogni problema di otimizzazione (come il TSP) ad un problema di meccanica statistica, interpretando la funzione costo $L(\vec{x})$ come l'energia $U$ del sistema e considerando la distribuzione di Boltzmann 

$$
p(\vec{x}) = \frac{e^{-\beta L(\vec{x})}}{Z}
$$

Allora si avrà che per $T \rightarrow 0$, cioè per $\beta \rightarrow \infty$, la distribuzione favorirà massimamente quegli $\vec{x}$ che minimizzano $L(\vec{x})$.

Per creare un algoritmo che faccia questo, si parte da un vettore $\vec{x}$ di interi che rappresenta un possibile percorso del commesso viaggiatore, e da una temperatura iniziale $T_0$, la quale va ridotta via via a piccoli step (di $ 0.01$ K o $ 0.001$ K ). Ad ogni valore di temperatura si eseguono $n$ step e a ogni step si propone una mossa verso un nuovo vettore di interi $\vec{x}'$. Questo nuovo vettore si otterrà con le mutazioni genetiche già presenti nel codice dell'algoritmo genetico dell'esercitazione 9, le quali ricordiamo essere:
- **Inversione** dell'ordine in cui compaiono $m < n$ elementi del vettore: il numero $m$ di elementi da invertire è estratto casualmente, così come il punto di inizio dell'inversione;
- **Permutazione di coppia** di due elementi del vettore estratti casualmente;
- **Permutazione di $m < n$ elementi** del vettore: il numero $m$ di elementi da permutare è estratto casualmente.

A questo punto la mossa proposta $\vec{x} \rightarrow \vec{x}' $ verrà accettata con probabilità $P$ definita nel seguente modo: 

- $P = 1$, se $L(\vec{x}')$ > $L(\vec{x})$ 
- $P = \exp \big[{-\beta \big( L(\vec{x}') - L(\vec{x})\big) \big]}$, se $L(\vec{x}')$ > $L(\vec{x})$ 

dove ricordiamo $\beta = \frac{1}{T}$.

L'algoritmo è stato implementato direttamente nel main, sfruttando in particolare quanto già fatto per l'esercitazione 9. Per il caso delle città lungo il cerchio si è partiti da una temperatura iniziale di $10$ K e si è proceduti verso $0$ riducendo ad ogni passo $T$ di $0.01$ K. Ad ogni temperatura si sono eseguiti poi $n = 100$ step dell'algoritmo. Per le città dentro il quadrato, invece, si è partiti sempre da una temperatura iniziale di $10$ K ma riducendola di $0.001$ K fino ad arrivare a $0$. Ad ogni temperatura vengono eseguiti $n = 300$ step.

Di seguito riportiamo i risultati ottenuti per le due simulazioni.

#### Città lungo un cerchio

```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

x1 = np.loadtxt("Route_00", unpack='true', usecols=0)
y1 = np.loadtxt("Route_00", unpack='true', usecols=1)
x2 = np.loadtxt("Route_fin", unpack='true', usecols=0)
y2 = np.loadtxt("Route_fin", unpack='true', usecols=1)
y3 = np.loadtxt("Media", unpack='true', usecols=1)
x3 = np.arange(y3.size)

fig = plt.figure(figsize=(20, 20))
fig.add_subplot(221)   #top left
plt.plot(x1, y1, marker='o')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('TSP: Initial Path', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

fig.add_subplot(222)  #top right
plt.plot(x2, y2, marker='o', c='g')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('TSP: Best Path', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

plt.figure(figsize=(10,7))
plt.plot(x3, y3, c='r')
plt.title('TSP: Path average', fontsize=22)
plt.xlabel('Step', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

plt.show();

print('Lunghezza percorso iniziale = ', y3[0])
print('Lunghezza percorso finale = ', y3[-1])
```

#### Città dentro un quadrato 

```python
x1 = np.loadtxt("Route_00_new", unpack='true', usecols=0)
y1 = np.loadtxt("Route_00_new", unpack='true', usecols=1)
x2 = np.loadtxt("Route_fin_new", unpack='true', usecols=0)
y2 = np.loadtxt("Route_fin_new", unpack='true', usecols=1)
y3 = np.loadtxt("Media_new", unpack='true', usecols=1)
x3 = np.arange(y3.size)

fig = plt.figure(figsize=(20, 20))
fig.add_subplot(221)   #top left
plt.plot(x1, y1, marker='o')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('TSP: Initial Path', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

fig.add_subplot(222)  #top right
plt.plot(x2, y2, marker='o', c='g')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('TSP: Best Path', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

plt.figure(figsize=(10,7))
plt.plot(x3, y3, c='r')
plt.title('TSP: Path average', fontsize=22)
plt.xlabel('Step', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

plt.show()

print('Lunghezza percorso iniziale = ', y3[0])
print('Lunghezza percorso finale = ', y3[-1])
```

Come si può vedere, entrambe le simulazioni convergono agli stessi percorsi trovati anche con l'algoritmo genetico. Tuttavia, è evidente che, nel caso del quadrato, questo richieda un tempo di calcolo notevole, dato che l'algoritmo esegue 300 step per ognuna delle 10000 temperature simulate. Inoltre, si segnala che per avere un miglior funzionamento dell'ottimizzazione, nel caso del quadrato si è eseguita due volte, all'interno dell'algoritmo, la permutazione su m elementi del vettore, per garantire una maggior varietà nella ricerca.

Rispetto all'algoritmo genetico utilizzato nell'esercitazione 9 si riscontrano varie differenze, in particolare nell'andamento della lunghezza del percorso al variare degli step dell'algoritmo. Questa infatti oscilla fortemente all'inizio, mantenendosi su valori piuttosto elevati, per poi crollare verso la fine della simulazione, quando la temperatura approccia lo zero. L'andamento è di fatto opposto a quanto visto per l'algoritmo genetico. Inoltre, dato l'elevato numero di operazioni svolte, questo tipo di algoritmo risulta meno efficiente rispetto a quello genetico.

### Esercizio 10.2

Lo scopo di questo secondo esercizio è quello di approcciare il problema del TSP tramite **parallelizzazione della computazione**. Questo verrà fatto sfruttando le librerie **MPICH**, che consentono di eseguire il codice su diversi core in modo parallelo e permettono di scambiare informazioni fra i diversi processi durante l'esecuzione.

L'esercizio richiede di eseguire il codice genetico sviluppato nell'esercitazione 9 su **4 diversi "continenti"**, eseguendo ricerche parallele, e facendo in modo che ogni $N_{migr}$ passi dell'algoritmo i continenti si scambino il migliore elemento della popolazione l'uno con l'altro. 

Per completare la richiesta si è deciso di utilizzare **due diversi generatori di numeri casuali**. Un generatore comune (my_rand0) a tutti e 4 i processi che realizzasse quei compiti che necessariamente dovevano essere uguali per tutti i continenti, ossia la generazione delle posizioni delle città da visitare e la creazione della popolazione di partenza, e uno invece inizializzato in modo diverso per ogni continente (my_rand) che generasse numeri casuali diversi per ogni nodo, il quale viene coinvolto all'interno dei processi di cross-over e di mutazione per le popolazioni dei singoli core. In questo modo tutti e 4 i nodi con cui si esegue il codice lavorano sull'ottimizzazione dello stesso problema ma generando popolazioni diverse gli uni dagli altri.

L'ultima aggiunta da fare riguarda lo **scambio del miglior elemento della popolazione fra continenti**. Questo scambio viene eseguito tramite le funzioni **MPI_send e MPI_recv**, le quali sono bloccanti per ogni singolo processo. Lo scambio avviene a coppie, cioè una coppia di continenti scambierà il proprio miglior elemento e l'altra coppia farà lo stesso. Per decidere quali continenti dovessero eseguire lo scambio si è utilizzato un vettore di 4 interi $(1, 2, 3, 4)$. Ogni volta in cui è necessario scegliere quali continenti debbano essere coinvolti nello scambio si esegue una permutazione a coppia del vettore per ottenere due coppie di continenti fra cui eseguire lo scambio. Ad esempio questa potrebbe dare la permutazione $(1, 4, 3, 2)$, la quale sta ad indicare che il continente 1 e 4 devono scambiarsi il loro miglior elemento, e la stessa cosa per i continenti 3 e 2. Dal momento che il vettore permutato deve essere uguale per tutti e quattro i processi, **questa permutazione viene eseguita tramite il generatore comune my_rand0**.

Di seguito riportiamo i risultati della parallelizzazione. Segnaliamo che ciascun continente lavora con popolazioni di $N = 600$ individui e scambia il miglior elemento ogni $N_{migr} = 25$ passi. 

```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

x = np.loadtxt("Route_00_new", unpack='true', usecols=0)
y = np.loadtxt("Route_00_new", unpack='true', usecols=1)
x1 = np.loadtxt("Route_quad_rk0", unpack='true', usecols=0)
y1 = np.loadtxt("Route_quad_rk0", unpack='true', usecols=1)
x2 = np.loadtxt("Route_quad_rk1", unpack='true', usecols=0)
y2 = np.loadtxt("Route_quad_rk1", unpack='true', usecols=1)
x3 = np.loadtxt("Route_quad_rk2", unpack='true', usecols=0)
y3 = np.loadtxt("Route_quad_rk2", unpack='true', usecols=1)
x4 = np.loadtxt("Route_quad_rk3", unpack='true', usecols=0)
y4 = np.loadtxt("Route_quad_rk3", unpack='true', usecols=1)
ym1 = np.loadtxt("Media_quad_rk0", unpack='true', usecols=0)
ym2 = np.loadtxt("Media_quad_rk1", unpack='true', usecols=0)
ym3 = np.loadtxt("Media_quad_rk2", unpack='true', usecols=0)
ym4 = np.loadtxt("Media_quad_rk3", unpack='true', usecols=0)
xm = np.arange(ym1.size)
ymb1 = np.loadtxt("Media_quad_rk0_best.dat", unpack='true', usecols=0)
ymb2 = np.loadtxt("Media_quad_rk1_best.dat", unpack='true', usecols=0)
ymb3 = np.loadtxt("Media_quad_rk2_best.dat", unpack='true', usecols=0)
ymb4 = np.loadtxt("Media_quad_rk3_best.dat", unpack='true', usecols=0)

#initial path
fig = plt.figure(figsize=(10, 7)) 
plt.plot(x, y, marker='o')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('TSP: Initial Path', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

#Final path
fig = plt.figure(figsize=(20, 20)) 
fig.add_subplot(221)   #top left: rank 0
plt.plot(x1, y1, marker='o')
plt.scatter(x1[0], y1[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('Final path: Rank 0', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

fig.add_subplot(222)  #top right: rank 1
plt.plot(x2, y2, marker='o')
plt.scatter(x2[0], y2[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('Final path : Rank 1', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

fig = plt.figure(figsize=(20, 20)) 
fig.add_subplot(221)   #bottom left: rank 2
plt.plot(x3, y3, marker='o')
plt.scatter(x3[0], y3[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('Final path: Rank 2', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

fig.add_subplot(222)  #bottom right: rank 3
plt.plot(x4, y4, marker='o')
plt.scatter(x4[0], y4[0], c='r', marker='*', s=500, label='Starting Point')
plt.title('Final path : Rank 3', fontsize=22)
plt.legend(loc='best', fontsize=15)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.grid(True)

#Path averages
fig = plt.figure(figsize=(20, 20)) 
fig.add_subplot(221)   #top left: rank 0
plt.plot(xm, ym1, label='Path average')
plt.plot(xm, ymb1, c='r', label='Best path')
plt.legend(loc='best', fontsize=15)
plt.title('Path average: Rank 0', fontsize=22)
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

fig.add_subplot(222)   #top right: rank 1
plt.plot(xm, ym2, label='Path average')
plt.plot(xm, ymb2, c='r', label='Best path')
plt.legend(loc='best', fontsize=15)
plt.title('Path average: Rank 1', fontsize=22)
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

fig = plt.figure(figsize=(20, 20)) 
fig.add_subplot(221)   #bottom left: rank 2
plt.plot(xm, ym3, label='Path average')
plt.plot(xm, ymb3, c='r', label='Best path')
plt.legend(loc='best', fontsize=15)
plt.title('Path average: Rank 2', fontsize=22)
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

fig.add_subplot(222)   #bottom right: rank 3
plt.plot(xm, ym4, label='Path average')
plt.plot(xm, ymb4, c='r', label='Best path')
plt.legend(loc='best', fontsize=15)
plt.title('Path average: Rank 3', fontsize=22)
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$L$', fontsize=18)
plt.grid(True)

plt.show();

#print('Percorso medio iniziale = ', y3[0])
#print('Percorso medio dopo 50 ripetizioni = ', y3[49])
#print('Percorso medio dopo 100 ripetizioni = ', y3[99])
#print('Percorso medio dopo 200 ripetizioni = ', y3[-1])
```

Questo rappresenta un notevole miglioramento rispetto al codice non parallelizzato, che convergeva al risultato esatto in circa 200 passi.
