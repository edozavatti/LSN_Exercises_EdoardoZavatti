```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 100
x = np.arange(N)  
L = 1000

x *=L 
f = np.loadtxt("media.dat", unpack='true')
error = np.loadtxt("errore_media.dat", unpack='true')
plt.errorbar(x, f, yerr=error)
plt.xlabel('Lanci')
plt.ylabel('<media> - 1/2')
plt.grid(True)
##plt.ylim([-0.008, 0.008])

plt.show();
```
