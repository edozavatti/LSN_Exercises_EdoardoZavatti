### Esercizio 3.1

Lo scopo di questo esercizio è di verificare numericamente la soluzione analitica data dalla teoria di Black-Scholes per il calcolo del fair price delle opzioni europee di tipo Call e Put. Questa teoria asserisce che l'andamento del prezzo segue un moto di tipo browniano geometrico (GBM), il quale ha come drift il tasso d'interesse "risk free" $r$ e come volatilità $\sigma$.

Un'**opzione** è un contratto in cui una parte si assume un onere mentre l'altra ha la possibilità di esercitare un diritto. Il tipo di opzione più semplice è quella **europea** in cui il contratto viene stipulato fra due parti. Il contratto prevede che chi vende l'opzione garantisca a chi la compra il diritto a comprare o a vendere un certo asset ad un prezzo pattuito $K$ in un tempo futuro $T$, tempo di scadenza del contratto. Dei due, **solo chi vende l'opzione ha l'obbligo di vendere o comprare l'asset al tempo T**, mentre l'altra parte può decidere di esercitare il proprio diritto sull'opzione solo se riesce a ricavarne un profitto. Questo avverrà se, al tempo $T$, il prezzo dell'asset sul mercato $S(T)$ è maggiore di quello pattuito $K$ al momento della stipulazione del contratto. In questo caso ($S(T) > K$) chi compra l'opzione eserciterà una **call option** e il suo profitto sarà dato da $max[0, S - K]$. Se invece il prezzo dell'asset è minore del prezzo pattuito ($S(T) < K$), chi compra l'opzione vorrà esercitare il proprio diritto a vendere l'asset, facendo quindi una **put option**. In questo secondo caso il profitto sarà dato da $max[0, K - S]$.

Ricordiamo che, per un prezzo di un asset che segue un andamento nel tempo dato da un moto browniano geometrico, cioè tale che  $S\sim GBM(\mu,\sigma^2)$, con valore al tempo $t = 0 $ pari a $S(0)$, il prezzo al tempo $ t $ sarà dato dalla seguente formula: 

$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$

con $W(t)\sim\mathcal{N}(0,t)$, ossia un numero distribuito secondo una gaussiana centrata in $0$ e con deviazione standard pari a $t$.
Inoltre, dal momento che gli incrementi di $W$ sono indipendenti e distribuiti in modo normale, è possibile discretizzare il moto senza commettere errori. Quindi possiamo dividere l'intervallo $\left[0, t \right]$ in $n$ sottointervalli temporali e campionare il prezzo dell'asset ai vari tempi $t_1 < t_2 < … < t_n$. Tale campionamento avvviene secondo la formula: 

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$

con $Z_i \sim \mathcal{N}(0,1)$.

L'esercizio richiede quindi di simulare il prezzo finale dell'asset $S(t)$ nei due modi, ossia dapprima campionando direttamente il risultato finale al tempo t, e in seguito discretizzando l'intervallo temporale e utilizzando il metodo ricorsivo. In entrambi i casi poi si richiede di calcolare i prezzi delle opzioni Call, $C[S(0),0]$, e Put, $P[S(0),0]$, con le relative incertezze. Per quanto spiegato all'inizio, questi saranno dati dalle seguenti formule:

$$ C[S(0),0] = e^{-rT} max(S - K) $$
$$ P[S(0),0] = e^{-rT} max(K - S) $$

Il fattore $e^{-rT}$ tiene conto dell'ipotetico tasso d'interesse che una banca avrebbe garatito al tempo $T$ se avessimo fatto un deposito al tempo iniziale $t_0 = 0$ al posto che investire sull'opzione.

Per la simulazione vengono usati i seguenti parametri:
- prezzo dell'asset a $t=0$: $S(0)=100$
- tempo di consegna: $T=1$
- prezzo pattuito: $K=100$
- tasso d'interesse risk-free: $r=0.1$
- volatilità: $\sigma=0.25$

La simulazione è stata fatta, sia nel caso diretto che discreto, eseguendo $10^5$ volte un moto browniano di $100$ passi, e suddividendo il tutto in $1000$ blocchi con i quali eseguire il data blocking per il calcolo delle incertezze. Il codice non richiede particolari spiegazioni, in quanto il tutto è stato eseguito in modo molto diretto nel main. 

Di seguito si riportano i 4 grafici per gli andamenti dei prezzi delle opzioni Call e Put al variare del numero di blocchi, ciascuno con relativa incertezza statistica. I grafici mostrano la differenza fra il valore simulato e il valore atteso dal calcolo analitico della teoria di Black-Scholes, che in questo caso è:

- Call = 14.975790778311286
- Put = 5.4595325819072364

Dunque, per essere compatibile con i valori attesi, i valori graficati devono essere compatibili con zero entro l'incertezza statistica.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

N = 1000
x = np.arange(N)
y = np.loadtxt("Call.dat", unpack='true')
error = np.loadtxt("errore_Call.dat", unpack='true')
plt.figure(figsize=(10,7))
plt.errorbar(x, y, yerr=error)
plt.ylabel(r'$C[S(0), 0] - 14.976$', fontsize=18)
plt.xlabel('Blocks', fontsize = 18)
plt.title('Single evolution: Call price', fontsize = 22)
plt.grid(True)

plt.show();

print('Risultato finale : ')
print(y[-1] + 14.975790778311286)
print('Incertezza statistica : ')
print(error[-1])
```
