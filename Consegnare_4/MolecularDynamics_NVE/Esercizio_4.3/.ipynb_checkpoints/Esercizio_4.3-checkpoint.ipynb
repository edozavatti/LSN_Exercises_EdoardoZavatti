# STATO SOLIDO

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

N = 9000

x1 = np.arange(N)
x2 = np.arange(N)
x3 = np.arange(N)
x4 = np.arange(N)
y = np.loadtxt("output_epot.dat", unpack='true')
z = np.loadtxt("output_etot.dat", unpack='true')
w = np.loadtxt("output_ekin.dat", unpack='true')
t = np.loadtxt("output_temp.dat", unpack='true')
#error = np.loadtxt("errore_Call.dat", unpack='true')

plt.axes([0., 0., 3., 3.])
plt.plot(x1, y)
plt.plot(x2, z)
plt.plot(x3, w)
plt.plot(x4, t)

plt.legend(["Energia potenziale", "Enegia totale", "Energia cinetica", "Temperatura"], loc ="best")

plt.xlabel('Passi')
plt.title('Plot energie stato solido (in unit Lennard-Jones)')
plt.grid(True)


plt.show()
```
