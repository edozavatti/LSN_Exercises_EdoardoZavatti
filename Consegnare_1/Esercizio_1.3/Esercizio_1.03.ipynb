```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M = 10000000
N = 100
L = int(M/N)

x = np.arange(N)
f = np.loadtxt("media_pi.dat", unpack='true')
error = np.loadtxt("errore_pi.dat", unpack='true')

x*=L

plt.errorbar(x, f, yerr=error)
plt.xlabel('Blocchi (5000 lanci per blocco)')
plt.ylabel('<pi_calcolato> - pi')
plt.title('Simulazione di Pi greco col metodo di Buffon')
plt.grid(True)


plt.show()
```
