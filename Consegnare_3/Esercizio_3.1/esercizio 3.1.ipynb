```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.optimize import curve_fit

M = 100000
N = 1000
k = 1

x = np.arange(N)
y = np.loadtxt("Call.dat", unpack='true')
error = np.loadtxt("errore_Call.dat", unpack='true')

plt.axes([0., 0., 2., 2.])
plt.errorbar(x, y, yerr=error)
plt.xlabel('Passi')

plt.title('Single evolution: Call price')
plt.grid(True)


plt.show()
```
