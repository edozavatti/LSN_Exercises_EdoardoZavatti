```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt
import scipy
#from scipy.optimize import curve_fit
from scipy.stats import norm

x = np.loadtxt("exp_1.dat", unpack='true')
y = np.loadtxt("exp_2.dat", unpack='true')
z = np.loadtxt("exp_10.dat", unpack='true')
w = np.loadtxt("exp_100.dat", unpack='true')

bins = np.linspace(0, 3, 100)

plt.hist(x, bins, alpha=0.5, density=True, label='N=1')
plt.hist(y, bins, alpha=0.5, density=True, label='N=2')
plt.hist(z, bins, alpha=0.5, density=True, label='N=10')
plt.hist(w, bins, alpha=0.5, density=True, label='N=100')

mu, sigma = scipy.stats.norm.fit(w)
best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)
plt.plot(bins, best_fit_line, color = "blue", label='fit gaussiano')

plt.legend(loc='upper right')

#plot
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Distribuzioni esponenziali')
plt.grid(True)

print('Fit gaussiano per N=100 :')
print()
print('media = ',  mu) 
print('deviazione standard = ', sigma) 



plt.show()
```

```python
import matplotlib as mlab
import numpy as np
import matplotlib.pyplot as plt
import scipy
#from scipy.optimize import curve_fit
from scipy.stats import cauchy

x = np.loadtxt("lorenz_1.dat", unpack='true')
y = np.loadtxt("lorenz_2.dat", unpack='true')
z = np.loadtxt("lorenz_10.dat", unpack='true')
w = np.loadtxt("lorenz_100.dat", unpack='true')

bins = np.linspace(-5, 5, 100)

plt.hist(x, bins, alpha=0.5, density=True, label='N=1')
plt.hist(y, bins, alpha=0.5, density=True, label='N=2')
plt.hist(z, bins, alpha=0.5, density=True, label='N=10')
plt.hist(w, bins, alpha=0.5, density=True, label='N=100')

mean, gamma = scipy.stats.cauchy.fit(w)
best_fit_line = scipy.stats.cauchy.pdf(bins, mean, gamma)
plt.plot(bins, best_fit_line, color = "blue", label='fit lorenziano')

plt.legend(loc='upper right')

#plot
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Distribuzioni lorenziane')
plt.grid(True)

print('Fit lorenziano per N=100 :')
print()
print('media = ',  mean) 
print('gamma = ', gamma) 


plt.show()
```
