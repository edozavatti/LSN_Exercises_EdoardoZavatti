### Esercizio 5.1

Lo scopo dell'esercizio è quello di implementare l'algoritmo di Metropolis e utilizzarlo per campionare la distribuzione di probabilità spaziale di un elettrone in un atomo di idrogeno. Infine, viene richiesto di servirsi di questa distribuzione per calcolare il valore medio di $\vec{r}$, dove quest'ultimo è il raggio vettore che collega il nucleo dell'atomo all'elettrone.

In meccanica quantistica, la funzione d'onda $\Psi$ è una funzione a valori complessi che descrive lo stato del sistema. In particolare, nel caso non relativistico di particella senza spin, la funzione d'onda $\Psi(\vec{r},t)$ può essere interpretata come un'ampiezza di probabilità, in modo tale che la sua norma quadratica, $|\Psi(\vec{r},t)|^2$, rappresenta la densità di probabilità di trovare il sistema (nel nostro caso l'elettrone) nella posizione $\vec{r}$ al tempo $t$.

Per un atomo ad idrogeno è possibile trovare in modo analitico la funzione d'onda del sistema. Essa è etichettata da tre numeri quantici (n, l, m) ed è indicata come $\Psi_{n,l,m}(r,\theta,\phi)$. Di seguito riportiamo le funzioni d'onda che ci interessano per risolvere l'esercizio, ossia la funzione d''onda di ground state $\Psi_{1,0,0}$ e quella di uno stato eccitato $2p$ $\Psi_{2,1,0}$:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$

$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$

Ricordiamo che $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm è il **raggio di Bohr**. Integrando analiticamente il quadrato di queste funzioni moltiplicato per $r$, è possibile trovare i seguenti valori attesi per i valori medi di posizione: 

$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

Come accennato sopra, per campionare queste distribuzioni di probabilità è stato richiesto di implementare l'algoritmo di Metropolis, il quale fornisce un modo sistematico ed efficiente per campionare distribuzioni di probabilità, anche qualora non se ne conoscesse in modo esplicito la normalizzazione, situazione piuttosto frequente in fisica statistica. Per implementare tale algoritmo è stato aggiunto un **metodo "Metropolis"** alla **classe "Posizione"**. Questa è parsa una scelta utile dal momento che quest'ultima classe era già equipaggiata con gli strumenti necessari per gestire, modificare e stampare le coordinate di un punto nello spazio 3D. Il metodo "Metropolis" prende in input la dimensione del passo da eseguire durante l'algoritmo e la distribuzione da campionare, ed esegue uno step dell'algoritmo. La distribuzione da campionare è passata al metodo sfruttando la **classe "FunzioneVettoriale"**, definita in **"FunzioneBase.h"**, che prende in ingresso una terna di coordinate e restituisce il valore (scalare) della funzione valutata in quel punto dello spazio. In questo modo il codice nel main risulta il più generico possibile, dando la possibilità di campionare distribuzioni diverse usando la stessa chiamata al metodo Metropolis.

L'algoritmo, in breve, genera un nuovo punto $\vec{x}$ a distanza fissata dal precedente punto $\vec{y}$ secondo una certa probabilità di transizione data da $T(\vec{x}|\vec{y})$. Questa "mossa" proposta dall'algoritmo viene accettata con probabilità $$\alpha = min\big[1,\frac{T(\vec{y}|\vec{x})p(\vec{x})}{T(\vec{x}|\vec{y})p(\vec{y})}\big]$$

In questo modo è garantito che i punti generati $\vec{x}$ siano distribuiti secondo la distribuzione di probabilità $p(\vec{x})$.

#### Equilibrazione 

L'algoritmo di Metropolis garantisce un buon campionamento dei punti solo asintoticamente. Ciò significa che, se partiamo da punti che inizialmente si trovano a grande distanza dal picco della distribuzione che vogliamo campionare, dovremo aspettare che l'algoritmo raggiunga l'equilibrio prima di iniziare a utilizzare i punti campionati per i calcoli che ci interessano. Nel nostro caso, andiamo a studiare quanto impiega l'algoritmo per raggiungere un buon equilibrio. 

Per vederlo, si è fatta partire la simulazione dal punto (x=15, y=0, z=0) e si è stampato passo per passo la distanza dei punti campionati dall'origine. Di seguito si riportano i risultati **in unità del raggio di Bohr $a_0$**. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

x1 = np.loadtxt("output_r_test.dat", unpack='true', usecols=0)
y = np.loadtxt("output_r_test.dat", unpack='true', usecols=1)
x2 = np.loadtxt("output_r_test1.dat", unpack='true', usecols=0)
z = np.loadtxt("output_r_test1.dat", unpack='true', usecols=1)

plt.figure(figsize=(8,5))
plt.plot(x1, y, c='b')
plt.xlabel('Step', fontsize=15)
plt.ylabel('$r/a_0$', fontsize=15)
plt.title('Equilibration for $\Psi_{1,0,0}$ ', fontsize=18)
plt.grid(True)

plt.figure(figsize=(8,5))
plt.plot(x2, z, c='r')
plt.xlabel('Step', fontsize=18)
plt.ylabel('$r/a_0$', fontsize=15)
plt.title('Equilibration for $\Psi_{2,1,0}$ ', fontsize=18)
plt.grid(True)

plt.show();
```

Come si può vedere, nel primo caso si raggiunge un buon equilibrio dopo circa 80 passi (in quanto da lì in poi i punti campionati sono a distanza $1.5 a_0$ dall'origine), mentre nel secondo caso dopo circa 130 passi (da lì in poi i punti sono nell'intorno di $5 a_0$). Un buon criterio per la simulazione, dunque sarebbe quello di cominciare ad effettuare le misure di posizione dopo circa 150 passi. In realtà, per le simulazioni successive, l'algoritmo è stato fatto partire da posizioni più ragionevoli, cioè a distanza 1.5 dall'origine per $\Psi_{1,0,0}$ e a distanza 5 per $\Psi_{2,1,0}$.

#### Regola empirica del 50% 

Per un buon campionamento dei dati, tali che i punti campionati non siano fortemente correlati fra loro ma che l'algoritmo non impieghi troppo tempo per generare nuove posizioni, è consigliato fare in modo che la probabilità di accettazione $\alpha$ sia circa uguale a 50%. Per controllare questo paramentro, è stato inserito nella classe "Posizione" un membro privato m_a che veniva opportunamente stampato alla fine della simulazione. Per operare su questo dato è necessario modificare la lunghezza $\delta$ del passo dell'algoritmo di Metropolis. Quando questo aumenta la probabilità di accettazione diminuisce, in quanto è più probabile generare punti in zone dello spazio a bassa probabilità, e viceversa.

Considerando una probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme dentro un cubo di lato $\delta$, si è trovato che i valori ottimali per questo parametro sono i seguenti: 
- Per $\Psi_{1,0,0}$: $\delta = 2.4$
- Per $\Psi_{1,0,0}$: $\delta = 3.0$

#### Calcolo di $\langle r \rangle$: probabilità di transizione uniforme

Ora che siamo in grado di campionare la distribuzione $|\Psi_{n, l, m}(\vec{r},t)|^2$, possiamo calcolare il valor medio di posizione, ossia l'integrale 
$$
\langle r \rangle = \int_0^\infty r |\Psi_{n, l, m}(\vec{r},t)|^2 r^2dr
$$ 
grazie ai metodi Monte Carlo. Per elaborare l'errore statistico sul calcolo, si sono suddivisi i $10^6$ passi dell'algoritmo in 100 blocchi. In generale, l'algoritmo di Metropolis tenderà a generare configurazioni correlate fra loro, in qanto ogni punto dipende strettamente dal precedente. Per poter considerare i blocchi come indipendenti l'uno dall'altro, quindi, è necessario eseguire un grande numero di passi Monte Carlo dentro ogni blocco. Nelle simulazioni successive si eseguono 1000 passi per blocco.

Di seguito si riportano i risultati di  $\langle r \rangle$ con relativo errore all'aumentare del numero di blocchi. **I risultati sono sempre in unità di raggio di Bohr**.

```python
N = 100 #numero di blocchi

x2 = np.arange(N)
z = np.loadtxt("output_r_ave.dat", unpack='true')
errorz = np.loadtxt("output_r_ave_error.dat", unpack='true')

plt.figure(figsize=(10,7))
plt.errorbar(x2, z - 1.5, yerr=errorz, c='b')

plt.xlabel('Blocchi', fontsize=18)
plt.ylabel(r'$\langle r/a_0 \rangle-\frac{3}{2}$', fontsize=18)
plt.title('$\Psi_{1,0,0}$', fontsize=22)
plt.grid(True)

plt.show()

print('Risultato finale : ')
print(z[-1])
print('Errore statistico : ')
print(errorz[-1])
```

#### Calcolo di $\langle r \rangle$: probabilità di transizione gaussiana

Le operazioni sopra descritte possono essere ripetute utilizzando una probabilità di transizione gaussiana. Anche in questo caso dobbiamo avere cura di equilibrare il sistema e di trovare un valore ottimale per il passo dell'algoritmo in modo da rispettare la regola empirica del 50%. In questo caso, questo passo $\delta$ non rappresenterà il lato del cubo entro cui stiamo generando punti uniformemente, ma bensì la varianza della gaussiana che usiamo come probabilità di transizione $T(\vec{x}|\vec{y})$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

x1 = np.loadtxt("output_r_testgauss.dat", unpack='true', usecols=0)
y = np.loadtxt("output_r_testgauss.dat", unpack='true', usecols=1)
x2 = np.loadtxt("output_r_testgauss1.dat", unpack='true', usecols=0)
z = np.loadtxt("output_r_testgauss1.dat", unpack='true', usecols=1)

plt.figure(figsize=(8,5))
plt.plot(x1, y, c='b')
plt.xlabel('Step', fontsize=15)
plt.ylabel('$r/a_0$', fontsize=15)
plt.title('Equilibration for $\Psi_{1,0,0}$ ', fontsize=18)
plt.grid(True)

plt.figure(figsize=(8,5))
plt.plot(x2, z, c='r')
plt.xlabel('Step', fontsize=18)
plt.ylabel('$r/a_0$', fontsize=15)
plt.title('Equilibration for $\Psi_{2,1,0}$ ', fontsize=18)
plt.grid(True)

plt.show()
```

In questo caso possiamo vedere che l'equilibrio si raggiunge, per entrambe le funzioni d'onda, dopo un centinaio di passi. I parametri ottimali trovati per la regola empirica del 50% sono:

- Per $\Psi_{1,0,0}$: $\delta = 0.75$
- Per $\Psi_{1,0,0}$: $\delta = 1.8$
