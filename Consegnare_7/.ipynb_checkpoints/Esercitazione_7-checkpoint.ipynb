Quest'analisi verrà applicata ad un codice di simulazione di dinamica molecolare fornitoci in laboratorio, "Monte Carlo NVT", il quale sfrutta l'algoritmo di Metropolis per eseguire misure di energia e pressione sul sistema. Inoltre, dal momento che il codice sfrutta l'algoritmo di Metropolis per campionare le distribuzioni delle molecole, sfruttiamo questa prima fase per trovare il parametro che garantisca un valore pari a 50% sull'accettazione. Si segnala, tuttavia, che grandi variazioni nel parametro $\delta$ della fase gassosa non portavano a grandi variazioni nel rate di accettazione, e che prendendo $\delta_{gas} = 5$ si otteneva un acceptance rate del 60%, dunque leggermente superiore a quanto volevamo.

Per rispondere alla prima domanda, l'esercizio chiede di sfruttare la funzione di autocorrelazione, la quale esprime il livello di correlazione esistente fra due variabili separate fra loro da $n$ passi. Essa è definita come segue: dato un insieme di $n_{max}$ campionamenti $x(m)$ con $m \in [1, n_{max}]$, la **funzione di autocorrelazione** è

$$
\chi(n) = \frac{\frac{1}{n_{max} - n} \sum_{n'=0}^{n_{max}-n}{x(n')x(n' + n)} - \frac{1}{n_{max} - n}\sum_{n'=0}^{n_{max}-n}{x(n')}\sum_{n'=0}^{n_{max}-n}{x(n' + n)}}{\frac{1}{n_{max}}\sum_{n'=0}^{n_{max}}{x^2(n')} - \big(\frac{1}{n_{max}}\sum_{n'=0}^{n_{max}}x(n')\big)^2}
$$

Dunque, $\chi(n)$ è una funzione del numero di step eseguiti $n$ e possiamo plottarla per le tre diverse fasi che vogliamo simulare.

### Esercizio 7.2 - 7.3

Lo scopo di questi esercizi è di implementare il calcolo della distribuzione radiale $g(r)$ in entrambi i codici di dinamica molecolare, quello utilizzato per il primo esercizio e quello utilizzato nell'esercitazione 4. 

Il calcolo della $g(r)$ viene eseguito dividendo l'intervallo $[0, L/2]$ (dove $L$ è il lato del box della simulazione) in **500 bin**. Valutando la distanza fra tutte le coppie di particelle, si riempie l'istogramma aumentando di $2$ il bin $r$ quando si trova una coppia di particelle a distanza $r + dr$ l'una dall'altra (essendo $dr$ la larghezza di un bin). Infine, bisogna aver cura di normalizzare l'istogramma, andando a dividere per il fattore $\rho N_{part}\frac{4\pi}{3} \big((r+dr)^3 - r^3\big)$. 

L'esercizio richiede di stampare sia il valore di $g(r)$ in ognuno dei blocchi utilizzati per la simulazione (in un file di output "output.gofr.0"), sia la media di g(r) fatta sui blocchi (nel file output.gave.0). Facciamo notare come, grazie al fatto che abbiamo utilizzato il metodo del data blocking anche per il riempimento dell'istogramma, siamo in grado di calcolare l'errore statistico anche sull'altezza di ogni bin.

### Esercizio 7.4

A questo punto siamo pronti per la simulazione vera e propria. Andiamo a eseguire il programma utilizzando **$100$ blocchi e $5000$ passi per blocco**, per un totale di $5\times10^5$ passi Monte Carlo. Di seguito riportiamo i risultati ottenuti per il calcolo di energia potenziale e pressione. Si segnala che, mentre la simulazione viene eseguita in unità di Lennard-Jones, i risultati sono presentati per un campione di **Argon** ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu). Infine, andiamo a stampare i risultati ottenuti per il grafico della distribuzione radiale g(r), sia quella ottenuta con il codice Molecular Dynamics NVE, sia quella ottenuta con il Monte Carlo NVT. Da ultimo è stato anche aggiunto un grafico che mostrasse la differenza fra le $g(r)$ campionate con i due diversi codici, per vedere se questa fosse compatibile con zero entro l'errore statistico dato dalla combinazione degli errori statistici delle due curve.

#### Fase Solida

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
import numpy as np
import math

########################################### DATI ARGON ######################################################

sigma = 0.34e-9 #angstrom
f_T = 120. # fattore moltiplicativo temperatura in Kelvin
m = 39.948 #amu
f_V = 86.9156461925 * 10**(-23.) #fattore moltiplicativo dell'energia potenziale in Joule
P0 = f_T / (sigma**3)

#############################################################################################################

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

N = 100 #blocks
x1 = np.arange(N)

y = np.loadtxt("output.epot_solidMC.0", unpack='true', usecols=(2))
y_err = np.loadtxt("output.epot_solidMC.0", unpack='true', usecols=(3))

plt.figure(figsize=(10,7))
plt.errorbar(x1, f_V*y, yerr=f_V*y_err)
plt.title('Potential energy per particle, solid phase', fontsize=22)
plt.xlabel('Blocks', fontsize=18)
plt.ylabel('U/N [J]', fontsize=18)
plt.grid(True)

z = np.loadtxt("output.pres_solidMC.0", unpack='true', usecols=(2))
z_err = np.loadtxt("output.pres_solidMC.0", unpack='true', usecols=(3))

plt.figure(figsize=(10,7))
plt.errorbar(x1, P0*z, yerr=P0*z_err, c='r')
plt.title('Pressure, solid phase', fontsize=22)
plt.xlabel('Blocks', fontsize=18)
plt.ylabel('P [Pa]', fontsize=18)
plt.grid(True)

r = np.loadtxt("output.gave_solid.0", unpack='true', usecols=(0))
r=r*sigma
g0 = np.loadtxt("output.gave_solid.0", unpack='true', usecols=(1))
g0_err = np.loadtxt("output.gave_solid.0", unpack='true', usecols=(2))
gMC = np.loadtxt("output.gave_solidMC.0", unpack='true', usecols=(1))
gMC_err = np.loadtxt("output.gave_solidMC.0", unpack='true', usecols=(2))

plt.figure(figsize=(10,7))
plt.errorbar(r, g0, yerr=g0_err, label="Molecular Dynamics")
plt.errorbar(r, gMC, yerr=gMC_err, label="MonteCarlo")
plt.title('Radial pair distribution, solid phase', fontsize=22)
plt.legend(loc ="best")
plt.xlabel('$r$ [m]', fontsize=18)
plt.ylabel('$g(r)$', fontsize=18)
plt.grid(True)

plt.figure(figsize=(10,7))
err_diff = np.sqrt(g0_err*g0_err + gMC_err*gMC_err)
plt.errorbar(r, g0 - gMC, yerr=err_diff)
#plt.errorbar(r, gMC, yerr=gMC_err, label="MonteCarlo")
plt.title('$g(r)$ comparison , solid phase', fontsize=22)
plt.xlabel('$r$ [m]', fontsize=18)
#plt.ylabel('$g(r)$', fontsize=18)
plt.grid(True)

plt.show()
```

Come si può vedere dai confronti fra le $g(r)$ , le due curve non sono sempre compatibili entro l'errore statistico. Questo può essere dovuto al fatto che si è diviso l'intervallo di campionamento della distribuzione in molti bin, e questo può aver ridotto l'errore statistico sul singolo bin. Ad ogni modo, il confronto fra le $g(r)$ ottenute nei due diversi tipi di simulazione mostra un buon accordo fra i sistemi, non stante siano due ensamble diversi (canonico e microcanonico) e benchè il tipo di campionamento delle posizioni nel tempo sia molto diverso fra essi. Infatti il codice Molecular Dynamics NVE campiona la dinamica molecolare nel tempo usando l'algoritmo di Verlet, mentre il Monte Carlo NVT sfrutta l'algoritmo di Metropolis.
