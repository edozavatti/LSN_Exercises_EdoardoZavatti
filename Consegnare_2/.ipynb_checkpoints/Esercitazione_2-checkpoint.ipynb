### Esercizio 2.1

In questo esercizio vogliamo calcolare, via metodi Monte Carlo, l'integrale:

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

Per farlo, prima di tutto useremo un metodo standard, generando uniformemente numeri in $\left[0, 1 \right]$, e, come secondo metodo, useremo il metodo dell'importance sampling campionando punti tramite una distribuzione lineare. La PDF utilizzata è la seguente: 

$$ P\left(x \right) = 2 \left(1 - x \right) $$

di cui riportiamo una breve visulizzazione qui sotto.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

x = np.linspace(0., 1., 100)
f = np.cos(math.pi/2. * x)
P = 2*(1 - x) 
plt.plot(x, math.pi/2. * f)
plt.plot(x, P)
plt.grid(True)
plt.show()
```

Come si vede dal grafico, la funzione lineare approssima meglio l'integranda rispetto alla distribuzione uniforme, quindi utilizzando il meteodo dell'importance sampling ci aspettiamo che l'integrale converga più velocemente e in modo più effeiciente al risultato atteso. 

Per risolvere l'esercizio si è sfruttata la classe **IntegralMC** definita in integralMC.cpp, ed in particolare si sono utilizzati i metodi **integralAVE** per eseguire il calcolo tramite campionamento uniforme e **integralAVE_par** per sfruttare l'importance sampling. Inoltre, per poter eseguire il calcolo su diverse integrande, entrambi i metodi prendono in input un puntatore a **FunzioneBase**, la quale è un'ulteriore classe che consente di scegliere fra diverse funzioni, ciascuna con un metodo **Eval** che restituisce il valore della funzione in un dato punto. Eval è un metodo virtuale, il funzionamento è definito in modo diverso in ciascuna funzione grazie al meccanismo dell'ereditarietà.

Riportiamo di seguito i grafici con l'andamento del valore dell'integrale all'aumentare del numero di blocchi. Per l'esercizio sono stati effettuati $N = 10^4 $ lanci Monte Carlo, suddivisi su $M = 100$ blocchi.

```python
M = 10000
N = 100
L = int(M/N)

x = np.arange(N)
f = np.loadtxt("media_I.dat", unpack='true')
error = np.loadtxt("errore_I.dat", unpack='true')


plt.figure(figsize=(10,7))
plt.errorbar(x, f, yerr=error)
plt.xlabel('Number of Blocks', fontsize=18)
plt.ylabel(r'$\langle I \rangle- 1$', fontsize=18)
plt.title('Integral - Uniform ', fontsize=22)
plt.grid(True)

plt.show()

print('Risultato finale : ')
print(f[-1] + 1.)
print('Incertezza : ')
print(error[-1])
```

```python
x = np.arange(N)
f = np.loadtxt("media_I_par.dat", unpack='true')
error = np.loadtxt("errore_I_par.dat", unpack='true')

plt.figure(figsize=(10,7))
plt.errorbar(x, f, yerr=error)
plt.xlabel('Number of Blocks', fontsize=18)
plt.ylabel(r'$\langle I \rangle- 1$', fontsize=18)
plt.title('Integral - Importance Sampling ', fontsize=22)
plt.grid(True)

plt.show()

print('Risultato finale : ')
print(f[-1] + 1.)
print('Incertezza : ')
print(error[-1])
```

Osserviamo come nel secondo caso l'andamento dell'integrale sia molto meno fluttuante e abbia un'incertezza statistica che ad ogni blocco è inferiore rispetto al caso del campionamento uniforme. Inoltre si nota come nel secondo caso l'integrale converga (entro l'errore statistico) dopo soli 60 blocchi al risultato esatto. Come ci aspettavamo, il metodo che sfrutta l'importance sampling risulta essere di maggior efficienza nell'esecuzione del calcolo.

### Esercizio 2.2

Lo scopo di questo esercizio è di simulare un random walk, dapprima in modo discreto muovendosi lungo una delle tre direzionei x, y, z all'interno di un reticolo cubico di costante $a = 1$, e poi nel continuo, eseguendo passi di lunghezza costante $a = 1$ in direzioni campionate uniformemente nell'angolo solido. L'obiettivo, per entrambi i casi, è di eseguire un plot dell'andamento di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$  (dove $\vec{r}_N$ è il vettore che congiunge l'origine con il punto raggiunto al passo $N$) e verificare che questo abbia un andamento del tipo $f(x) = k\sqrt{N}$, con $k$ una costante da determinare.

Per il caso discreto, il random walk è implementato tramite l'estrazione di un numero casuale, il quale serve a determinare in quale delle tre direzioni eseguire il passo e se ci si muove in avanti e indietro. Per il caso continuo, invece, ci si serve di due numeri per campionare una direzione nell'angolo solido: gli angoli $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$. Per l'implementazione del codice ci si è serviti della classe **Posizione**, la quale permette di immagazzinare le tre coordinate cartesiane $(x, y, z)$ e restituisce i valori delle coordinate di altri sistemi di riferimento (in particolare coordinate sferiche e cilindriche).

L'esercizio prevede la ripetizione di $10^4$ random walks ciascuno di 100 passi. Il calcolo dell'errore statistico è realizzato grazie al data blocking suddividendo i random walks in 100 blocchi. Di seguito vengono mostrati i grafici per il caso continuo e discreto dell'andamento di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ al variare del numero $N$ di passi. Per entrambi i grafici è stato possibile fare un fit con la funzione $f(x) = k\sqrt{N}$.   

```python
from scipy.optimize import curve_fit

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

M = 10000
N = 101
k = 1

x = np.arange(N)
y = np.loadtxt("media_random_walk.dat", unpack='true')
error = np.loadtxt("errore_random_walk.dat", unpack='true')

plt.figure(figsize=(10,7))
plt.errorbar(x, y, yerr=error, fmt='.', color='red')
plt.xlabel('Steps', fontsize=18)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle}$', fontsize=18)
plt.title('Discrete random walk', fontsize=22)
plt.grid(True) 

p_opt, p_cov = curve_fit(f, x, y, bounds=([0],[2]))
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit, color='b', label = 'fit $k\sqrt{N}$') # plotting fitted function
plt.legend(loc='best', fontsize=15)

print("Risultati del fit : ")
print("Parametro ottimizzato k =", p_opt)
print('Incertezza del parametro =',  np.sqrt(np.diagonal(p_cov)))
print()

plt.show()

print("Valor medio della distanza dall'origine al passo 100: ", y[99],  'pm'  , error[99])
```

```python
z = np.zeros(N)
plt.figure(figsize=(10,7))
plt.errorbar(x, y - p_opt[0] * np.sqrt(x), yerr=error, label="simulated pi")

plt.plot(x, z, color = "red")
plt.xlabel('Steps', fontsize=18)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle}$ - $k\sqrt{N}$', fontsize=18)
plt.title('Discrete random walk: compatibility test', fontsize=22)
plt.grid(True)

plt.show()
```

Questo secondo grafico evidenzia la differenza fra il fit dato dalla funzione $k\sqrt{N}$ e il plot di $\sqrt{\langle |\vec{r}_N|^2 \rangle}$. Come si può vedere, i due grafici sono quasi ovunque compatibili, in quanto la loro differenza è quasi ovunque zero entro l'incertezza statistica.

```python
N = 100
x = np.arange(N)
y = np.loadtxt("media_random_walk_continuo.dat", unpack='true')
error = np.loadtxt("errore_random_walk_continuo.dat", unpack='true')

plt.figure(figsize=(10,7))
plt.errorbar(x, y, yerr=error, fmt='.', color='red')
plt.xlabel('Steps', fontsize=18)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle}$', fontsize=18)
plt.title('Continuous random walk', fontsize=22)
plt.grid(True) 

p_opt, p_cov = curve_fit(f, x, y, bounds=([0.8],[1.2]))
y_fit = f(x,p_opt[0])
plt.plot(x,y_fit, color = 'b', label = 'fit $k\sqrt{N}$') # plotting fitted function
plt.legend(loc='best')

print("Parametro ottimizzato [k] =", p_opt)
print("Incertezza del parametro =", np.sqrt(np.diagonal(p_cov)))

plt.show();

print("Valor medio della distanza dall'origine al passo 100: ", y[N-1], " pm " , error[N-1])
```

```python
z = np.zeros(N)
plt.figure(figsize=(10,7))
p_opt, p_cov = curve_fit(f, x, y, bounds=([0.8],[1.2]))
plt.errorbar(x, y - p_opt[0] * np.sqrt(x), yerr=error, label="simulated pi")
plt.plot(x, z, color = "red")
plt.xlabel('Passi', fontsize=18)
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle}$ - $k\sqrt{N}$', fontsize=18)
plt.title('Continuous random walk: compatibility test', fontsize=22)
plt.grid(True)

plt.show()
```

Anche nel caso continuo si evidenzia una buona compatibilità fra il fit $k\sqrt{N}$ e il plot di $\sqrt{\langle |\vec{r}_N|^2 \rangle}$.
