### Esercizio 1.1

Lo scopo del primo esercizio è quello di valutare il corretto funzionamento del generatore di numeri pseudo-casuali fornito a noi studenti in laboratorio.
Una prima verifica viene effettuanta andando a calcolare l'integrale $\langle r \rangle = \int_0^1 r dr$ tramite tecniche Monte Carlo e verificando che il risultato sia $ 1/2 $. Questo è infatti il risultato che ci aspettiamo quando calcoliamo il valore medio di numeri generati in modo uniforme fra $0$ e $1$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

N = 100
x = np.arange(N)  
L = 1000

x *=L 
f = np.loadtxt("media.dat", unpack='true')
plt.figure(figsize=(10,7))
error = np.loadtxt("errore_media.dat", unpack='true')
plt.errorbar(x, f, yerr=error)
plt.xlabel(r'Number of Throws', fontsize=20)
plt.ylabel(r'$\langle r \rangle-\frac{1}{2}$', fontsize=20)
plt.title('Mean', fontsize=20)
plt.grid(True)

plt.show();

print('Risultato finale : ')
print(f[-1] + 0.5)
print('Errore statistico : ')
print(error[-1])
```

Come si può vedere, il risultato converge correttamente al valore atteso, entro l'errore statistico. Quest'ultimo è stato calcolato con il metodo del data blocking: sono stati effettuati $ M = 1000000 $ lanci Monte Carlo suddivisi su $ N = 100 $ blocchi. Per ogni blocco $i$ sono stati calcolati i valori $ r_i $ e $ r^2_i $, per $ i = 1, 2, ... , N $. A questo punto, il valore medio $\langle r \rangle $ viene ottenuto come 

$$ \quad \langle r\rangle \simeq \frac{1}{N} \sum_{i=1}^N r_i $$

e la deviazione standard $$\sigma^2_r := \langle (r-\langle r\rangle)^2\rangle = \langle r^2 \rangle -\langle r\rangle^2 $$

è calcolata usando le forumule

 $$ \langle r^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N r^2_i \quad \quad \langle r\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N r_i \right)^2 $$
 
 Infine, la deviazione standard dalla media, graficata in figura tramite le barre d'errore, è calcolata come: 
 
 $$\frac{\sigma}{\sqrt{N}} \simeq \sqrt{\frac{1}{N} \left[ \frac{1}{N} \sum_{i=1}^N r^2_i - \left( \frac{1}{N} \sum_{i=1}^N r_i \right)^2 \right]} $$ .

Come si può notare dal grafico, l'incertezza decresce all'aumentare del numero di blocchi coinvolti nel calcolo statistico, e il risultato converge ad un valore sempre più preciso.

Come secondo test, siamo interessati a calcolare la varianza dei numeri generati uniformemente tra $ 0 $ e $ 1 $ e vogliamo verificare che questa sia $ 1/12 $. Per farlo calcoliamo l'integrale  $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 100
x = np.arange(N)  
L = 1000

x *=L 
plt.figure(figsize=(10,7))
f = np.loadtxt("varianza.dat", unpack='true')
error = np.loadtxt("errore_varianza.dat", unpack='true')
plt.errorbar(x, f, yerr=error)
plt.xlabel(r'Number of Throws', fontsize=20)
plt.ylabel(r'$\sigma^2 - \frac{1}{12}$', fontsize=20)
plt.grid(True)
plt.title('Variance', fontsize=20)

plt.show();

print('Risultato finale : ')
print(f[-1] + 1/12)
print('Errore statistico : ')
print(error[-1])
```

Anche in questo caso il calcolo converge al risultato atteso entro l'errore statistico. Quest'ultimo è stato calcolato ancora con il metodo del data blocking, nella stessa maniera descritta in precedenza.

**Test del $\chi^2$**

Come ultimo test vogliamo verificare la probabilità che i punti che stiamo generando siano effettivamente distribuiti in maniera uniforme. Per farlo, utilizziamo un test del $\chi^2$. Dividiamo l'intervallo $\left[0, 1 \right]$ in $ 100 $ sottointervalli uguali ed eseguiamo 100 volte il test, ciascuno con $n = 10^4$ numeri pseudo-casuali. Eseguiamo quindi un plot di $\chi^2_j$ per $j = 1, 2, ..., 100$ così definito: 
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$

dove $n_i$ sono il numero di punti compresi fra $i \times \frac{1}{100}$ e $(i + 1) \times \frac{1}{100}$, per $i = 0, 1, 2, ..., 99$.

```python
N = 100
x = np.arange(N) 

plt.figure(figsize=(10,7))
f = np.loadtxt("Chi_2.dat", unpack='true')
plt.plot(x, f)
plt.xlabel(r'\#Test', fontsize=20)
plt.ylabel(r'$\chi^2$', fontsize=20)
plt.title('Test del $\chi^2$', fontsize=20)
plt.grid(True)
plt.show();
```

Qualora l'ipotesi che i numeri fossero estratti da una distribuzione uniforme fosse corretta, ci aspetteremmo un valore di $\chi^2$ che approssima il numero di sotto-intervalli in cui abbiamo diviso $\left[0, 1 \right]$, ossia nel nostro caso ci aspetteremmo $\chi^2 \simeq 100$. Come si può vedere dal grafico, però, il test fornisce risultati che si discostano, talvolta anche per più di 30 unità, dal valore atteso. Questo è un forte indice che l'ipotesi (numeri estratti da una distribuzione uniforme) sia improbabile. Effettivamente i numeri sono solo psudo-casuali, in quanto estratti non realmente a caso ma tramite un algoritmo deterministico.

### Esercizio 1.2

In questo esercizio ci proponiamo di verificare il teorema del limite centrale (TLC), il quale asseirisce che al crescere di $N$, la quantità $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ tende a distribuirsi come una gaussiana, qualunque sia la distribuzione di probabilità $p(x_i)$ dalla quale vengono estratti gli $ x_i $, a patto che $p(x_i)$ abbia varianza finita. 
L'esercizio richiede quindi di plottare l'evoluzione di $S_N$ per tre diverse distribuzioni (**uniforme, esponenziale, lorenziana**), al variare di $ N = 1, 2, 10, 100 $. 

Per verificare se il teorema è sodisfatto o meno eseguiamo anche un fit dell'ultimo istogramma con una gaussiana, della quale riportiamo, sotto al grafico, il valor medio $\mu$ e la varianza $\sigma$.

**Distribuzione Esponenziale**

```python
import scipy
#from scipy.optimize import curve_fit
from scipy.stats import norm

x = np.loadtxt("exp_1.dat", unpack='true')
y = np.loadtxt("exp_2.dat", unpack='true')
z = np.loadtxt("exp_10.dat", unpack='true')
w = np.loadtxt("exp_100.dat", unpack='true')

bins = np.linspace(0, 3, 100)
plt.figure(figsize=(10,7))
plt.hist(x, bins, alpha=0.5, density=True, label='N=1')
plt.hist(y, bins, alpha=0.5, density=True, label='N=2')
plt.hist(z, bins, alpha=0.5, density=True, label='N=10')
plt.hist(w, bins, alpha=0.5, density=True, label='N=100')

mu, sigma = scipy.stats.norm.fit(w)
best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)
plt.plot(bins, best_fit_line, color = "blue", label='Gaussian fit')

plt.legend(loc='upper right', fontsize=15)

#plot
plt.xlabel('x', fontsize=18)
plt.ylabel('Probability', fontsize=18)
plt.title('Exponential distribution', fontsize=22)
plt.grid(True)

plt.show()

print('Fit gaussiano per N=100 :')
print()
print('media = ',  mu) 
print('deviazione standard = ', sigma) 
```

**Distribuzione Uniforme**

```python
x = np.loadtxt("uniforme_1.dat", unpack='true')
y = np.loadtxt("uniforme_2.dat", unpack='true')
z = np.loadtxt("uniforme_10.dat", unpack='true')
w = np.loadtxt("uniforme_100.dat", unpack='true')

bins = np.linspace(0, 1, 100)
plt.figure(figsize=(10,7))
plt.hist(x, bins, alpha=0.5, density=True, label='N=1')
plt.hist(y, bins, alpha=0.5, density=True, label='N=2')
plt.hist(z, bins, alpha=0.5, density=True, label='N=10')
plt.hist(w, bins, alpha=0.5, density=True, label='N=100')

mu, sigma = scipy.stats.norm.fit(w)
best_fit_line = scipy.stats.norm.pdf(bins, mu, sigma)
plt.plot(bins, best_fit_line, color = "blue", label='Gaussian fit')

plt.legend(loc='upper right', fontsize=15)

#plot
plt.xlabel('x', fontsize=22)
plt.ylabel('Probability', fontsize=18)
plt.title('Uniform distribution', fontsize=18)
plt.grid(True)

plt.show();

print('Fit gaussiano per N=100 :')
print()
print('media = ',  mu) 
print('deviazione standard = ', sigma) 
```

**Distribuzione Lorenziana**

```python
from scipy.stats import cauchy

x = np.loadtxt("lorenz_1.dat", unpack='true')
y = np.loadtxt("lorenz_2.dat", unpack='true')
z = np.loadtxt("lorenz_10.dat", unpack='true')
w = np.loadtxt("lorenz_100.dat", unpack='true')

bins = np.linspace(-5, 5, 100)
plt.figure(figsize=(10,7))
plt.hist(x, bins, alpha=0.5, density=True, label='N=1')
plt.hist(y, bins, alpha=0.5, density=True, label='N=2')
plt.hist(z, bins, alpha=0.5, density=True, label='N=10')
plt.hist(w, bins, alpha=0.5, density=True, label='N=100')

mean, gamma = scipy.stats.cauchy.fit(w)
best_fit_line = scipy.stats.cauchy.pdf(bins, mean, gamma)
plt.plot(bins, best_fit_line, color = "blue", label='Lorentzian fit')

plt.legend(loc='upper right', fontsize=15)

#plot
plt.xlabel('x', fontsize=18)
plt.ylabel('Probability', fontsize=18)
plt.title('Lorentzian distribution', fontsize=22)
plt.grid(True)

plt.show()

print('Fit lorenziano per N=100 :')
print()
print('media = ',  mean) 
print('gamma = ', gamma) 
```

Mentre nei primi due casi è evidente come le distribuzioni varino all'aumentare di $N$ per approcciare una gaussiana, nell'ultimo istogramma si può notare come la distribuzione lorenziana non cambi in modo sostanziale. Infatti, quest'ultima distribuzione non è a varianza finita e quindi non segue il TLC. Inoltre, il bacino di attrazione della distribuzione lorenziana è la lorenziana stessa, e per questo non si notano grandi variazioni all'aumentare di $N$. In quest'ultimo caso il fit eseguito non è con una gaussiana ma con una lorenziana.

### Esercizio 1.3

Lo scopo di quest'ultimo esercizio è quello di simulare l'esperimento dell'ago di Buffon per calcolare il valore di $\pi$. Questo esperimento prevede di lanciare un ago di lunghezza $l$ sopra una griglia di linee equispaziate a distanza $d$. La probabilità che l'ago intercetti una linea è pari a $P = 2L/\pi d$, dalla quale, una volta calcolata la probabilità come rapporto fra $N_{lanci}$ e $N_{intercetti}$, è possibile ricavare $\pi$.

Per svolgere il compito, si è deciso di effettuare $10^7$ lanci Monte Carlo suddivisi su 100 blocchi. Ad ogni lancio corrisponde la generazione, fatta tramite il generatore di numeri casuali, della posizione del centro dell'ago nell'intervallo $\left[0, d/2 \right]$, e di un angolo in $\left[0, \pi \right]$. Per generare quest'ultimo numero senza utilizzare in modo esplicito il numero $\pi$ si è utilizzato il metodo accept/reject, il quale consiste nel generare una posizione $\left(x, y \right) $ all'interno di un rettangolo ($x\in [-1;+1]$, $y\in [0;+1]$), e solo nel caso in cui la distanza di questo punto 2D dal centro sia minore di 1 si ritorna il valore dell'angolo fra il punto e l'asse x tramite la formula 
$$\theta = \cos^{-1}\left({\frac{x}{\sqrt{x^2 + y^2}}}\right)$$
Una volta generati la posizione $c$ del centro e l'inclinazione $\theta$ dell'ago, la condizione per cui la riga viene toccata dall'ago è data da $ c < \sin \left(\theta \right) $. Solo quando la disuguaglianza è soddisfatta, viene aumentato il contatore dei successi $N_{intercetti}$. Infine il valore di $\pi$ viene calcolato come: 

$$\pi = \frac{2L}{Pd} = \lim_{N_{lanci} \to \infty}\frac{2L}{d}\frac{N_{lanci}}{N_{intercetti}}$$

Con questa formula viene eseguita una stima di $\pi$ per ognuno dei 100 blocchi. Poi, tramite il data blocking, si calcolano valore medio ed incertezza statistica. Si riportano di seguito i risultati della simulazione.

```python
import math
M = 10000000
N = 100
L = int(M/N)

x = np.arange(N)
f = np.loadtxt("media_pi_new.dat", unpack='true')
error = np.loadtxt("errore_pi_new.dat", unpack='true')
x*=L
plt.figure(figsize=(10,7))
plt.errorbar(x, f, yerr=error)
plt.xlabel('Throws', fontsize=20)
plt.ylabel(r'$\langle \pi_{calculated} \rangle-\pi$', fontsize=20)
plt.title('Simulazione di Pi greco col metodo di Buffon', fontsize=20)
plt.grid(True)

plt.show()

print('Risultato finale : ')
print(f[-1] + math.pi)
print('Errore statistico : ')
print(error[-1])
```
