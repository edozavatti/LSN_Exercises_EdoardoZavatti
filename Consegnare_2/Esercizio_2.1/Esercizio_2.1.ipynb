```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M = 10000
N = 100
L = int(M/N)

x = np.arange(N)
f = np.loadtxt("media_I.dat", unpack='true')
error = np.loadtxt("errore_I.dat", unpack='true')

x*=L

plt.errorbar(x, f, yerr=error)
plt.xlabel('Lanci')
plt.ylabel('<Integrale> - 1')
plt.title('Integrale con metodo Monte Carlo')
plt.grid(True)


plt.show();
```
